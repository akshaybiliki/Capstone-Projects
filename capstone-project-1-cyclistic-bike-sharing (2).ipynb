# ANALYSIS AND SHARE

```r

# postiom= dogde - To create Grouped Bar Chart

all_months_v2 %>% 
  mutate(day_of_week= wday(started_at, label = TRUE)) %>% 
  group_by(member_casual,month,day_of_week) %>% 
  summarise(number_of_rides=n(), mean_duration = mean(ride_length2/60)) %>% 
  arrange(member_casual,day_of_week) %>% 
  ggplot(aes(x=day_of_week, y= mean_duration,fill= member_casual))+ geom_col(position = 'dodge')+
  labs(title = "Average Ride Duration by Day and Rider Type")+ylab("Average Duration (minutes)") + 
  xlab("Day of Week")
```
